```python
import csv

# Open the CSV file
with open('car.data', 'r') as f:
    car_data = csv.reader(f, delimiter='\n')

    # Iterate through the CSV rows
    for i in car_data:
        # Split each row into values
        values = []
        for val in i[0].split(','):
            values.append(f"'{val}'")

        # Construct and execute the SQL query
        query = f"INSERT INTO cardatasheet.car VALUES ({', '.join(values)})"
        cursor.execute(query)

# Commit the changes to the database
mydb.commit()
```
